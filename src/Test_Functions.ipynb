```python
from random import randrange
import pandas as pd
import string
import random
```

```python

def random_score_txt(train_txt, train_labels_txt):
    corpus = open(train_txt, encoding="utf-8")
    true_labels = open(train_labels_txt, encoding="utf-8")
    docs_labels = [int(doc.strip()) for doc in true_labels.readlines()]
    docs = [doc.strip() for doc in corpus.readlines()]
    predicted_label = [randrange(2) for i in docs]
    df = pd.DataFrame([docs,docs_labels,predicted_label]).T
    df.columns = ['Text','True_Label', 'Predicted_Label']
    return df
```

```python
random_score_txt("../datasets/agnews/train.txt","../datasets/agnews/train_labels.txt")
```

```python
def random_cate_vocab(word, size = 50):
    cate_vocab = []
    occurences = []
    for i in range(size):
        new_word = word
        for j in range(randrange(7)):
            new_word += random.choice(string.ascii_letters)
        cate_vocab.append(new_word)
        occurences.append(randrange(10,10000))
    df = pd.DataFrame([cate_vocab, occurences]).T
    df.columns = ['Vocab', 'Occurences']
    df = (df.sort_values('Occurences', ascending = False)).reset_index(drop = True)
    return df
        
```

```python
random_cate_vocab('damn')
```
